##5. Посчитайте экономический эффект от каждой инициативы

Параметры:
- Коэффициент дисконтирования (r) = 15% годовых
- Горизонт расчета — 3 года, чтобы виден был эффект с учётом внедрения и поддержки
- Средний доход с пользователя - 500 руб./мес.
- Среднее количество пользователей - 1000.
- Затраты на тех.поддержку - 300 час./мес. * 500 руб./час. = 300 000 руб. * 12 = 1 800 000 руб.
---

1) Оптимизация платежной системы и управления подписками  

- Предположения:  
  - Снижение оттока из-за платежных проблем на 5% (от 1000 пользователей = 50 пользователей)    
  - Дополнительный доход в месяц = 25 000 руб.
  - В год = 300 000 руб.
  - Затраты на внедрение и поддержку: 300 000 руб. в первый год

Расчет:  
Год 0: -300 000 (затраты)  
Годы 1-3: +300 000  

NPV = -300 000 + 300 000 / (1+0.15)^1 + 300 000 / (1+0.15)^2 + 300 000 / (1+0.15)^3  
NPV = 384 967 руб.  
IRR = 84%  
Срок окупаемости = 1 год

---
2) Автоматизация обработки обратной связи и поддержки учеников через чат-боты  

 - Предположения:  
  - Снижает время обработки на 30%, то есть затраты на техподдержку снижаются на 90 час/мес, что дает экономию 540 000 руб. в год.
  - Затраты на разработку чат-бота — 150 000 руб. (однократно)  
  - Рост удовлетворенности и удержания на 1% (10 пользователей) — дополнительный доход 60 000 в год  
  - Затраты на поддержку чат-бота - 500 000 руб./год.  

Расчет:  
Год 0: -150 000  
Годы 1-3: 540 000 + 60 000 - 500 000 = +100 000 руб./год

NPV = -150 000 + 100 000 / (1+0.15)^1 + 100 000 / (1+0.15)^2 + 100 000 / (1+0.15)^3  
NPV = 78 322 руб.  
IRR = 45%
Срок окупаемости 1,5 года

---
3) Внедрение на образовательную платформу раздела «Поговорим с психологом»  
Предположения:  
- Рост удовлетворенности и удержания на 7% (70 пользователей) — дополнительный доход 420 000 в год
- Дополнительные затраты (поддержка): 300 000 в год  
- Затраты на внедрение (в т.ч. разработка ИИ-модели): 500 000 в первый год  

Расчет:  
Год 0: -500 000  
Годы 1-3: 420 000 - 300 000 = 120 000 руб/год

NPV = -500 000 + 120 000 / (1+0.15)^1 + 120 000 / (1+0.15)^2 + 120 000 / (1+0.15)^3  
NPV = 226 012 руб.  
IRR = 15%  
Срок окупаемости > 4 лет

- Экономия на работе специалистов: 500 000 в год
